```python
import pandas as pd
import numpy as np
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return 0

ratings_movies = pd.read_csv('data/ratings_movies.csv')
# Выделяем год в отдельную колонку
ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
ratings_movies['year_release'] = ratings_movies['year_release'].astype(int)
ratings_movies.info()
# 8.2 Фильм 1999 года с наименьшим рейтингом    
pivot = ratings_movies[ratings_movies['year_release'] == 1999].pivot_table(
    values='rating', index=['movieId','title'], aggfunc='mean', fill_value=0
)
display(pivot.sort_values(by='rating',ascending=True))

# сочетание жанров фильмов (genres), выпущенных в 2010 году, получило наименьшую среднюю оценку (rating)
pivot = ratings_movies[ratings_movies['year_release'] == 2010].pivot_table(
    values='rating', index='genres', aggfunc='mean', fill_value=0
)
#display(pivot.sort_values(by='rating',ascending=True))

# Какой пользователь (userId) посмотрел наибольшее количество различных (уникальных) жанров (genres) фильмов
#UserId = 0
#max_genres = 0
#for user in ratings_movies['userId']: 
#    counts1 = len(ratings_movies[ratings_movies['userId'] == user]['genres'].unique())
#    if counts1 > max_genres:
#        max_genres = counts1
#        UserId = user
#print (UserId, max_genres)
# Найдите пользователя, который выставил наименьшее количество оценок, но его средняя оценка фильмам наибольшая.
display (ratings_movies.groupby('userId')['rating'].agg(['count', 'mean']).sort_values(by=['count','mean'], ascending=[True,False]))

"""Найдите сочетание жанров (genres) за 2018 году, которое имеет наибольший средний рейтинг (среднее по столбцу rating), 
и при этом число выставленных ему оценок (количество значений в столбце rating) больше 10."""
pivot = ratings_movies[ratings_movies['year_release'] == 2018].pivot_table(
    values='rating', index=['genres'], aggfunc=['mean','count']
)
display(pivot.sort_values(by=[('mean','rating'),('count','rating')],ascending=[False,True]))
```
