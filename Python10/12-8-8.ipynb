```python
import pandas as pd
import numpy as np
import re
orders = pd.read_csv('data/orders.csv', delimiter=';')
#orders.info()
products = pd.read_csv('data/products.csv', delimiter=';')
#products['Product_ID'] = products['Product_ID'].astype(int)
#products.info()
orders_products = orders.merge(products,left_on = 'ID товара', right_on = 'Product_ID', how = 'left')
#orders_products['Product_ID'] = orders_products['Product_ID'].astype(np.int64)
orders_products.info()
lost_orders = []
for i in orders['ID товара']:
    if i not in list(products['Product_ID']):
        lost_orders.append(i)
#print (orders)
display(orders_products)
orders_products['Summ'] = orders_products['Price']*orders_products['Количество'] 
mask = (orders_products['Статус'] == 'Оплачен, формируется к отправке')|(orders_products['Статус'] == 'Выполнен')
pivot = orders_products[mask].pivot_table(
    values = 'Summ',
    index = 'ID Покупателя',
    aggfunc = 'sum')
display (pivot.sort_values(by='Summ',ascending=False))
```
